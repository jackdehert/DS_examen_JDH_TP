Alex thinks that there are two laptops, such that the price of the first laptop is less (strictly smaller) than the price of the second laptop but the quality of the first laptop is higher (strictly greater) than the quality of the second laptop.

Please, check the guess of Alex. You are given descriptions of n laptops. Determine whether two described above laptops exist.

Input

The first line contains an integer n (1 ≤ n ≤ 105) — the number of laptops.

Next n lines contain two integers each, ai and bi (1 ≤ ai, bi ≤ n), where ai is the price of the i-th laptop, and bi is the number that represents the quality of the i-th laptop (the larger the number is, the higher is the quality).

All ai are distinct. All bi are distinct.

Output

If Alex is correct, print "Happy Alex", otherwise print "Poor Alex" (without the quotes).

```python
## dummy input
2
1 2
2 1

2
2 2
3 1
```

```python
n = int(input())  
lijst = []


for i in range(n):
    x, y = map(int, input().split())
    lijst.append((x, y))


if all(x == y for x, y in lijst):
    print("Poor Alex")
else:
    
    lijst.sort(key=lambda tup: (tup[0], -tup[1]))

  
    for i in range(1, n):
        x1, y1 = lijst[i-1]
        x2, y2 = lijst[i]
        
        
        if x1 < x2 and y1 > y2:
            print("Happy Alex")
            break
    else:
        print("Poor Alex")
```

## [466A - Cheap Travel - Difficulty 1200 ](https://codeforces.com/problemset/problem/466/A)


#### Put in dummy input

```python
# 6 2 1 2

# 5 2 2 3

# 10 3 5 1

# 101 110 1 100
```

```python
planned_rides, special_ticket_rides, single_price, special_price = map(int, input().split())

single_tickets = planned_rides * single_price
only_special_tickets = ((planned_rides + special_ticket_rides - 1) // special_ticket_rides ) * special_price
required_special_tickets = (planned_rides // special_ticket_rides) * special_price
remaining_single_tickets = (planned_rides % special_ticket_rides) * single_price
combination_tickets = required_special_tickets + remaining_single_tickets

result = min(single_tickets, only_special_tickets, combination_tickets)
print(result)
```

## [1999B - Card Game- Difficulty 1000 +3 ](https://codeforces.com/problemset/problem/1999/B)


#### Put in dummy input

```python
# 5
# 3 8 2 6
# 1 1 1 1
# 10 10 2 2
# 1 1 10 10
# 3 8 7 2
```

```python
tests = int(input())

results = []

for test in range(tests):
    a1, a2, b1, b2 = map(int,input().split())
    suneet_wins = 0

    suneet_hands = [(a1,a2), (a2, a1)]
    slavic_hands = [(b1,b2), (b2,b1)]

    for suneet_hand in suneet_hands:
        for slavic_hand in slavic_hands:
            suneet_rounds = 0
            slavic_rounds = 0

            for suneet_card, slavic_card in zip(suneet_hand, slavic_hand):
                if suneet_card > slavic_card:
                    suneet_rounds += 1
                elif suneet_card < slavic_card:
                    slavic_rounds += 1
            if suneet_rounds > slavic_rounds:
                suneet_wins += 1
    results.append(suneet_wins)
for result in results:
    print(result)
```
