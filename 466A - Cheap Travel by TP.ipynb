## [466A - Cheap Travel - Difficulty 1200 ](https://codeforces.com/problemset/problem/466/A)


#### Put in dummy input

```python
# 6 2 1 2

# 5 2 2 3

# 10 3 5 1

# 101 110 1 100
```

```python
planned_rides, special_ticket_rides, single_price, special_price = map(int, input().split())

single_tickets = planned_rides * single_price
only_special_tickets = ((planned_rides + special_ticket_rides - 1) // special_ticket_rides ) * special_price
required_special_tickets = (planned_rides // special_ticket_rides) * special_price
remaining_single_tickets = (planned_rides % special_ticket_rides) * single_price
combination_tickets = required_special_tickets + remaining_single_tickets

result = min(single_tickets, only_special_tickets, combination_tickets)
print(result)
```
