```python
# 43A - Football

n = int(input())  # Read the number of lines
first_team = ""
second_team = ""
first_team_count = 0
second_team_count = 0

for goal in range(n):
    team = input()
    if first_team == "":  # First goal sets the first team
        first_team = team
        first_team_count += 1
    elif team == first_team:  # Increment first team's count if they score
        first_team_count += 1
    elif team != first_team:
        second_team = team
        second_team_count += 1

# Determine the winner
if first_team_count > second_team_count:
    print(first_team)
else:
    print(second_team)
```
