## [785A - Anton and Polyhedrons - Difficulty 800 ](https://codeforces.com/problemset/problem/785/A)


#### Put in dummy input

```python
# 4
# Icosahedron
# Cube
# Tetrahedron
# Dodecahedron
```

```python
n= int(input())
number_of_faces = 0

for i in range(n):
    item = input()
    if item == "Tetrahedron":
        number_of_faces += 4
    elif item == "Cube":
        number_of_faces += 6
    elif item == "Octahedron":
        number_of_faces += 8
    elif item == "Dodecahedron":
        number_of_faces += 12
    else:
        number_of_faces += 20
print(number_of_faces)
```
