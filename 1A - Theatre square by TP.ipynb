## [1A - Theatre square - Difficulty 1000](https://codeforces.com/problemset/problem/1/A)




#### Put in dummy input

```python
# 6 6 4
# 1 1 1
# 2 1 1
```

```python
n,m,a = map(int,input().split())

if n % a == 0: 
    r1 = n // a 
else : 
    r1 = (n // a) + 1 
if m % a == 0: 
    r2 = m // a 
else : 
    r2 = (m // a) + 1
print(int(r1 * r2))
```
