```python
#122A Lucky Division

n = int(input())

def is_lucky(number):
    """Check if a number is a lucky number (contains only digits 4 and 7)."""
    for digit in str(number):
        if digit not in "47":
            return False
    return True

def is_almost_lucky(n):
    """Check if n is almost lucky (divisible by a lucky number)."""
    for i in range(1, n + 1):
        if is_lucky(i) and n % i == 0:
            print("YES")
            return  # Stop further execution once a lucky divisor is found
    print("NO")  # Print "NO" if no lucky divisor is found

is_almost_lucky(n)
```
